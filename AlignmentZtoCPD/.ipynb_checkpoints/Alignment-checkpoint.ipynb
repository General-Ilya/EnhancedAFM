```python
import glob
import numpy as np
import matplotlib.pyplot as plt
import cv2

import re

def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    (See Toothy's implementation in the comments)
    '''
    return [ atoi(c) for c in re.split('(\d+)', text) ]
```

```python
filelist = glob.glob('figs/Z/*.tiff')
filelist.sort(key=natural_keys)
print(filelist)
Z_array = np.array([np.array(cv2.imread(fname, 0)) for fname in filelist])
```

```python
filelist = glob.glob('figs/CPD/*.tiff')
filelist.sort(key=natural_keys)
print(filelist)
CPD_array = np.array([np.array(cv2.imread(fname, 0)) for fname in filelist])
```

```python
transformation = cv2.estimateRigidTransform(Z_array[0], Z_array[1], 0)
transformation
dx = np.zeros_like(transformation)
dx[0,2] = transformation[0,2]
```

```python
x = cv2.warpAffine(Z_array[0], dx, Z_array[0].T.shape)
y = cv2.addWeighted(x, 0.5, Z_array[1], 0.5,0)
plt.figure(figsize=(20,20))
plt.imshow(y)
plt.show()
```
